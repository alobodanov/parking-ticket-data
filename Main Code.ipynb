```python
csv_tickets_2 = "parkingTicketData/Resources/parking_tickets_2018/Parking_Tags_Data_2018_2.csv"
ticket_df_2 = pd.read_csv(csv_tickets_2)
ticket_df_2.head()
```

```python
csv_tickets_3 = "parkingTicketData/Resources/parking_tickets_2018/Parking_Tags_Data_2018_3.csv"
ticket_df_3 = pd.read_csv(csv_tickets_3)
ticket_df_3.head()
```

```python
tickets_2018_df=pd.concat([ticket_df_1,ticket_df_2,ticket_df_3])
locations = list(tickets_2018_df["location2"])

```

```python
#store clean addresses only
clean_locations = []
for location in locations:
    try:
        if(re.search(r'\d+\w+', location)):
            
            clean_locations.append(location)
    except:
        continue
        

```

```python
clean_locations2 = clean_locations[933495:]
```

```python
clean_locations2 = list(dict.fromkeys(clean_locations2))
```

```python
clean_locations = clean_locations1 + clean_locations2
clean_locations = list(dict.fromkeys(clean_locations))
```
